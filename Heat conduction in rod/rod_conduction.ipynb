# Heat Conduction in Rod Mathematical Modeling

## 1. Introduction to Heat

Heat is a form of energy that transfers from one body to another due to a temperature difference. At the molecular level, heat represents the kinetic energy of randomly moving particles. Particles in a hot region have higher kinetic energy than those in a cooler region, thus energy flows from hot to cold until thermal equilibrium is reached.

### Key Concepts:
- **Temperature (T)**: A measure of the average kinetic energy of particles
- **Heat (Q)**: Energy in transit due to temperature difference
- **Thermal Energy**: The total kinetic energy of all particles in a system
- **Heat Flux (q)**: Rate of heat transfer per unit area (W/m²)

## 2. Modes of Heat Transfer

Heat can be transferred through three fundamental mechanisms:

### 2.1 Conduction
Heat transfer through direct molecular contact within a solid or between solids in contact. Energy transfers from high-energy particles to low-energy particles through collisions and interactions at the atomic level.

### 2.2 Convection
Heat transfer in fluids (liquids and gases) involving bulk motion of the fluid. Includes:
- **Natural convection**: Driven by density differences due to temperature variations
- **Forced convection**: Driven by external forces (fans, pumps)

### 2.3 Radiation
Heat transfer through electromagnetic waves, requiring no medium. All objects emit thermal radiation based on their temperature.

## 3. Heat Conduction

### 3.1 Physical Mechanism
In conduction, heat flows through a material without bulk motion of the material itself. At the atomic level:
- In metals: Free electrons carry energy
- In non-metals: Lattice vibrations (phonons) transfer energy
- Energy flows from regions of high molecular kinetic energy to low molecular kinetic energy

### 3.2 Fourier's Law of Heat Conduction

  <img src="Resources\fourier_potrait.webp" alt="RC Discharging Circuit" width = "200" style="margin-right: 20px;"/>

**One-Dimensional Form:**
$$
q = -k × \frac{dT}{dx}
$$

Where:
- q = heat flux (W/m²)
- k = thermal conductivity (W/m·K)
- dT/dx = temperature gradient (K/m)
- The negative sign indicates heat flows from hot to cold

**Three-Dimensional Vector Form:**

$$q⃗ = -k∇T$$

Where ∇T is the temperature gradient vector.

### 3.3 Physical Interpretation
- **Thermal Conductivity (k)**: Material property indicating ability to conduct heat
  - High k: Good conductor (metals: k = 50-400 W/m·K)
  - Low k: Good insulator (air: k ≈ 0.026 W/m·K)
- **Temperature Gradient**: Driving force for heat conduction
- **Heat Flux**: Proportional to temperature gradient

## 4. Derivation of the Heat Equation from First Principles

### 4.1 Starting Point: Conservation of Energy

We begin with the fundamental principle that energy cannot be created or destroyed, only converted from one form to another.

**Energy Balance for a Control Volume:**

'''
Rate of energy in - Rate of energy out + Rate of energy generated = Rate of energy stored
'''

### 4.2 Consider a One-Dimensional Rod

  <img src="Resources\heat-conduction-through-a-rod-l.jpg" alt="RC Discharging Circuit" width = "400" style="margin-right: 20px;"/>

</div>

**Step 1: Define the Control Volume**
Consider a small element of thickness dx at position x:
- Left face at position x
- Right face at position x + dx
- Cross-sectional area A

### 4.3 Apply Fourier's Law

**Heat flux entering at x:**
$$
q_{in} = -k × \frac{∂T}{∂x}|_x
$$

**Heat flux leaving at x + dx:**
$$
q_{out} = -k × \frac{∂T}{∂x}|_{x+dx}
$$

Using Taylor expansion:
$$
q_{out} = -k × \frac{∂T}{∂x}|_x - k × \frac{∂²T}{∂x²} × dx
$$

**Net heat flow into the element:**
$$
q_{net} = q_{in} - q_{out} = k × \frac{∂²T}{∂x²} × dx
$$

**Rate of heat flow into the element:**
$$
Q̇_{net} = A × q_{net} = k × A × \frac{∂²T}{∂x²} × dx
$$

### 4.4 Energy Storage Term

The rate of energy storage equals the rate of change of internal energy.

For a solid with:
- Density: ρ (kg/m³)
- Specific heat: c (J/kg·K)
- Volume of element: A × dx

**Rate of energy storage:**
$$
Q̇_{stored} = ρ × c × A × dx × \frac{∂T}{∂t}
$$

### 4.5 Heat Generation Term

If there's internal heat generation (e.g., electrical heating, nuclear reactions):
$$
Q̇_{generated} = q̇ × A × dx
$$

Where q̇ is the volumetric heat generation rate (W/m³).

### 4.6 Complete Energy Balance

Substituting into the energy balance equation:
$$
k × A × \frac{∂²T}{∂x²} × dx + q̇ × A × dx = ρ × c × A × dx × \frac{∂T}{∂t}
$$

Dividing by A × dx:
$$
k × \frac{∂²T}{∂x²} + q̇ = ρ × c × \frac{∂T}{∂t}
$$

### 4.7 The One-Dimensional Heat Equation

**General form with heat generation:**
$$
\frac{∂T}{∂t} = α × \frac{∂²T}{∂x²} + \frac{q̇}{ρc}
$$

**Without heat generation (q̇ = 0):**
$$
\frac{∂T}{∂t} = α × \frac{∂²T}{∂x²}
$$

Where **α = k/(ρc)** is the **thermal diffusivity** (m²/s).

### 4.8 Three-Dimensional Heat Equation

Extending to three dimensions:
$$
\frac{∂T}{∂t} = α × \left(\frac{∂²T}{∂x²} + \frac{∂²T}{∂y²} + \frac{∂²T}{∂z²}\right) + \frac{q̇}{ρc}
$$

Or using the Laplacian operator ∇²:
$$
\frac{∂T}{∂t} = α∇²T + \frac{q̇}{ρc}
$$

## 5. Key Assumptions in the Derivation

### 5.1 Material Properties
1. **Constant thermal conductivity (k)**: Independent of temperature and position
2. **Constant density (ρ)**: No thermal expansion
3. **Constant specific heat (c)**: Independent of temperature
4. **Isotropic material**: Same properties in all directions

### 5.2 Physical Assumptions
1. **Continuum assumption**: Material treated as continuous medium
2. **Local thermal equilibrium**: Temperature well-defined at each point
3. **No convection**: Heat transfer only by conduction
4. **No radiation**: Negligible radiative heat transfer
5. **One-dimensional heat flow**: Temperature varies only with x

### 5.3 Mathematical Assumptions
1. **Smooth temperature distribution**: Temperature is differentiable
2. **Linear temperature profile**: Valid for small dx in Taylor expansion
3. **Fourier's law validity**: Linear relationship between heat flux and gradient

## 6. Sources of Error in Real Scenarios

### 6.1 Material Property Variations
**Reality vs. Assumption:**
- Thermal conductivity often depends on temperature: k = k(T)
- Density changes with temperature due to thermal expansion
- Specific heat varies with temperature and phase changes

**Error Impact:** Can lead to 10-50% errors in temperature predictions

### 6.2 Non-uniform Material Properties
**Reality vs. Assumption:**
- Real materials may have impurities, grain boundaries, or composite structures
- Property variations in space: k = k(x,y,z)

**Error Impact:** Local hot spots or temperature variations not captured

### 6.3 Multi-dimensional Effects
**Reality vs. Assumption:**
- Real objects are 3D, not 1D
- Heat loss/gain from sides, ends, surfaces

**Error Impact:** Faster cooling/heating than predicted, non-uniform temperature profiles

### 6.4 Boundary Conditions
**Reality vs. Assumption:**
- Perfect thermal contact rarely exists
- Contact resistance at interfaces
- Convective and radiative losses at surfaces

**Error Impact:** Temperature drops at interfaces, different surface temperatures

### 6.5 Transient Effects
**Reality vs. Assumption:**
- Finite heat propagation speed (very small effect)
- Non-local effects in very rapid heating

**Error Impact:** Usually negligible except in ultrafast heating scenarios

### 6.6 Scale Effects
**Reality vs. Assumption:**
- Continuum assumption breaks down at molecular scales
- Quantum effects in nanostructures
- Ballistic heat transport in thin films

**Error Impact:** Reduced effective thermal conductivity in nanostructures

### 7.2 Time Scales
For a rod of length L, the characteristic time for heat diffusion is:
$$
τ = L²/α
$$

**Examples:**
- Copper rod (L = 0.1 m): τ ≈ 89 seconds
- Steel rod (L = 0.1 m): τ ≈ 12 minutes
- Glass rod (L = 0.1 m): τ ≈ 37 hours

## 8. Building the Mathematical Model: Design Decisions

With the theoretical foundation established, we must make specific decisions to create a workable model. Let's address each key question:

### 8.1 Define Specific Geometry and Boundary Conditions

**Decision: Simple 1D Rod with Fixed End Temperatures**

**Geometry:**
- Rod length: L = 1.0 m
- Uniform cross-section (area cancels out in 1D equation)
- Material: Steel (representative metal with moderate thermal diffusivity)

**Boundary Conditions (Dirichlet - Fixed Temperature):**
- Left end (x = 0): T = 100°C (hot end)
- Right end (x = L): T = 0°C (cold end, could represent ice water)

**Justification:**
- Fixed temperature boundaries are the simplest to implement mathematically
- Represents practical scenarios like one end in hot water, other in ice bath
- Eliminates complexity of convective heat transfer coefficients
- Creates clear temperature gradient for visualization

### 8.2 Choose Solution Method

**Decision: Explicit Finite Difference Method**

**Numerical Parameters:**
- Spatial discretization: 51 points (Δx = 0.02 m)
- Time step: Δt = 0.1 s
- Mesh Fourier number: r = αΔt/Δx² ≈ 0.35 (< 0.5 for stability)

**Justification:**
- Explicit method is simpler to understand and implement
- No need to solve linear systems at each time step
- Stability condition is easily satisfied with chosen parameters
- Computational efficiency adequate for 1D problem

### 8.3 Implement Initial Conditions

**Decision: Uniform Initial Temperature**

**Initial Condition:**
- T(x,0) = 20°C (room temperature throughout rod)

**Justification:**
- Physically realistic: rod initially at room temperature
- Creates maximum initial temperature gradients at boundaries
- Simple to implement mathematically
- Demonstrates clear heat diffusion from both ends
- Easy to visualize temperature evolution

### 8.4 Material Properties

**Decision: Steel Properties**

**Properties:**
- Thermal conductivity: k = 50 W/m·K
- Density: ρ = 7850 kg/m³
- Specific heat: c = 450 J/kg·K
- Thermal diffusivity: α = k/(ρc) = 1.42×10⁻⁵ m²/s

**Justification:**
- Steel is common engineering material
- Moderate thermal diffusivity shows clear but not instantaneous heat transfer
- Well-documented properties in literature
- Creates realistic time scales (minutes to hours for full heating)

### 8.5 Validation Strategy

**Approach:**
1. **Steady-State Check:** Verify final temperature profile is linear
2. **Energy Conservation:** Monitor total system energy
3. **Boundary Condition Verification:** Ensure boundaries remain fixed
4. **Physical Reasonableness:** Heat flows from hot to cold, monotonic approach to steady state

**Expected Behavior:**
- Initial rapid temperature change near boundaries
- Gradual heating of rod center
- Eventual linear steady-state profile: T(x) = 100 - 100x/L
- Time constant τ ≈ L²/α ≈ 70,000 seconds ≈ 19.4 hours

### 8.6 Model Limitations and Error Sources

**Acknowledged Simplifications:**
1. **1D Assumption:** Ignores radial temperature variations
2. **Constant Properties:** Real steel properties vary with temperature
3. **Perfect Boundary Conditions:** Real boundaries have thermal resistance
4. **No Heat Loss:** Ignores convection/radiation to surroundings
5. **Continuum Assumption:** Valid for macroscopic rod dimensions

**Expected Errors:**
- Numerical truncation: O(Δx²) + O(Δt) ≈ 0.1% for chosen discretization
- Physical modeling: 5-15% due to property variations and heat losses
- Boundary condition idealization: 2-10% depending on actual boundary setup


*These design decisions create a simple yet realistic model that balances mathematical tractability with physical meaning, suitable for educational demonstration and basic engineering analysis.*

```python
"""
Animated Heat Conduction Model in a 1D Rod
==========================================

This script demonstrates heat conduction in a 1D rod using the explicit finite difference method.
The animation shows how temperature distributes over time from fixed boundary conditions.

Based on the heat equation: ∂T/∂t = α ∂²T/∂x²
Where α is thermal diffusivity = k/(ρc)

Author: Heat Transfer Modeling Example
Date: 2025
"""

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.colors import LinearSegmentedColormap

# ============================================================================
# PHYSICAL PARAMETERS
# ============================================================================

# Rod geometry
L = 1.0                    # Rod length [m]
 
# Material properties (Steel)
k = 50.0                   # Thermal conductivity [W/m·K]
rho = 7850.0              # Density [kg/m³]
c = 450.0                 # Specific heat [J/kg·K]
alpha = k / (rho * c)     # Thermal diffusivity [m²/s]

print(f"Material Properties (Steel):")
print(f"Thermal conductivity: {k} W/m·K")
print(f"Density: {rho} kg/m³")
print(f"Specific heat: {c} J/kg·K")
print(f"Thermal diffusivity: {alpha:.2e} m²/s")

# Boundary conditions (Dirichlet - fixed temperatures)
T_left = 100.0            # Left boundary temperature [°C]
T_right = 0.0             # Right boundary temperature [°C]
T_initial = 20.0          # Initial temperature throughout rod [°C]

# ============================================================================
# NUMERICAL PARAMETERS
# ============================================================================

# Spatial discretization
nx = 51                   # Number of spatial grid points
dx = L / (nx - 1)         # Spatial step size [m]
x = np.linspace(0, L, nx) # Spatial grid

# Temporal discretization
dt = 0.1                  # Time step [s]
r = alpha * dt / dx**2    # Mesh Fourier number (stability parameter)

# Check numerical stability (explicit method requires r ≤ 0.5)
print(f"\nNumerical Parameters:")
print(f"Spatial points: {nx}")
print(f"Spatial step size: {dx:.4f} m")
print(f"Time step: {dt} s")
print(f"Mesh Fourier number: {r:.4f}")

if r > 0.5:
    print("WARNING: Unstable numerical scheme! Reduce dt or increase nx.")
    print("Continuing anyway for demonstration...")
else:
    print("✓ Stable numerical scheme")

# Animation parameters
total_time = 3600         # Total simulation time [s] (1 hour)
nt = int(total_time / dt) # Number of time steps
frames_to_show = 200      # Number of animation frames
frame_interval = max(1, nt // frames_to_show)  # Show every nth time step

print(f"Total simulation time: {total_time} s ({total_time/3600:.1f} hours)")
print(f"Total time steps: {nt}")
print(f"Animation frames: {frames_to_show}")

# ============================================================================
# INITIALIZE TEMPERATURE FIELD
# ============================================================================

# Create temperature array
T = np.ones(nx) * T_initial    # Initialize to room temperature
T[0] = T_left                  # Apply left boundary condition
T[-1] = T_right                # Apply right boundary condition

# Store temperature history for animation
T_history = []
time_history = []

# Calculate steady-state solution for reference
T_steady = T_left + (T_right - T_left) * x / L

print(f"\nInitial Conditions:")
print(f"Left boundary: {T_left}°C")
print(f"Right boundary: {T_right}°C")
print(f"Initial temperature: {T_initial}°C")

# ============================================================================
# FINITE DIFFERENCE TIME STEPPING
# ============================================================================

print("\nRunning simulation...")

for n in range(nt):
    # Store current state for animation (only at specified intervals)
    if n % frame_interval == 0:
        T_history.append(T.copy())
        time_history.append(n * dt)
    
    # Create temporary array for new temperatures
    T_new = T.copy()
    
    # Apply finite difference scheme to interior points
    # T[i]^(n+1) = T[i]^n + r*(T[i+1]^n - 2*T[i]^n + T[i-1]^n)
    for i in range(1, nx-1):
        T_new[i] = T[i] + r * (T[i+1] - 2*T[i] + T[i-1])
    
    # Update temperature field
    T = T_new
    
    # Apply boundary conditions (these remain fixed)
    T[0] = T_left
    T[-1] = T_right
    
    # Progress indicator
    if n % (nt // 10) == 0:
        progress = 100 * n / nt
        print(f"Progress: {progress:.0f}%")

# Add final state
T_history.append(T.copy())
time_history.append(total_time)

print("✓ Simulation completed!")

# ============================================================================
# CREATE ANIMATED VISUALIZATION
# ============================================================================

print("Creating animation...")

# Set up the figure and axis
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))
fig.suptitle('Heat Conduction in a 1D Steel Rod', fontsize=16, fontweight='bold')

# Color map for temperature visualization
colors = ['blue', 'cyan', 'green', 'yellow', 'orange', 'red']
cmap = LinearSegmentedColormap.from_list('temperature', colors)

# Initialize plots
line1, = ax1.plot([], [], 'b-', linewidth=3, label='Current Temperature')
line_steady, = ax1.plot(x, T_steady, 'r--', linewidth=2, alpha=0.7, label='Steady State')
line_initial, = ax1.plot(x, np.ones(nx) * T_initial, 'g:', linewidth=2, alpha=0.7, label='Initial Temperature')

# Temperature profile plot setup
ax1.set_xlim(0, L)
ax1.set_ylim(-5, 105)
ax1.set_xlabel('Position along rod [m]')
ax1.set_ylabel('Temperature [°C]')
ax1.set_title('Temperature Distribution')
ax1.grid(True, alpha=0.3)
ax1.legend()

# Color-coded temperature visualization
im = ax2.imshow([[T_initial] * len(x)], aspect='auto', cmap=cmap, vmin=0, vmax=100,
                extent=[0, L, -0.1, 0.1])
ax2.set_xlim(0, L)
ax2.set_ylim(-0.1, 0.1)
ax2.set_xlabel('Position along rod [m]')
ax2.set_ylabel('Rod Cross-section')
ax2.set_title('Temperature Color Map')

# Add colorbar
cbar = plt.colorbar(im, ax=ax2, orientation='horizontal', pad=0.2)
cbar.set_label('Temperature [°C]')

# Text annotations for current time and key information
time_text = ax1.text(0.02, 0.98, '', transform=ax1.transAxes, fontsize=12,
                    verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8))

info_text = ax1.text(0.98, 0.02, f'Steel Rod\nL = {L} m\nα = {alpha:.2e} m²/s', 
                    transform=ax1.transAxes, fontsize=10, verticalalignment='bottom',
                    horizontalalignment='right', bbox=dict(boxstyle='round', facecolor='lightblue', alpha=0.8))

def animate(frame):
    """Animation function called for each frame"""
    
    # Get current temperature distribution
    T_current = T_history[frame]
    current_time = time_history[frame]
    
    # Update temperature profile line
    line1.set_data(x, T_current)
    
    # Update color map
    T_2d = T_current.reshape(1, -1)  # Reshape for imshow
    im.set_array(T_2d)
    
    # Update time display
    time_hours = current_time / 3600
    time_text.set_text(f'Time: {current_time:.0f} s ({time_hours:.2f} h)')
    
    # Calculate and display current statistics
    T_center = T_current[nx//2]  # Temperature at center
    max_gradient = np.abs(np.gradient(T_current, dx)).max()  # Maximum temperature gradient
    
    # Update info text with current stats
    stats_text = (f'Steel Rod\nL = {L} m\nα = {alpha:.2e} m²/s\n\n'
                 f'Center Temp: {T_center:.1f}°C\nMax Gradient: {max_gradient:.1f}°C/m')
    info_text.set_text(stats_text)
    
    return line1, im, time_text, info_text

# Create animation
anim = animation.FuncAnimation(fig, animate, frames=len(T_history), 
                             interval=50, blit=False, repeat=True)

# Adjust layout
plt.tight_layout()

# ============================================================================
# DISPLAY RESULTS AND ANALYSIS
# ============================================================================

print("\n" + "="*60)
print("SIMULATION RESULTS AND ANALYSIS")
print("="*60)

# Calculate theoretical time constant
tau_theoretical = L**2 / alpha
print(f"Theoretical time constant: {tau_theoretical:.0f} s ({tau_theoretical/3600:.1f} hours)")

# Analyze final temperature distribution
T_final = T_history[-1]
T_center_final = T_final[nx//2]
T_steady_center = T_steady[nx//2]

print(f"\nFinal Results:")
print(f"Center temperature (final): {T_center_final:.2f}°C")
print(f"Center temperature (steady): {T_steady_center:.2f}°C")
print(f"Approach to steady state: {100*T_center_final/T_steady_center:.1f}%")

# Energy analysis
initial_energy = np.sum(np.ones(nx) * T_initial) * dx
final_energy = np.sum(T_final) * dx
steady_energy = np.sum(T_steady) * dx

print(f"\nEnergy Analysis:")
print(f"Initial total thermal energy (relative): {initial_energy:.2f}")
print(f"Final total thermal energy (relative): {final_energy:.2f}")
print(f"Steady-state thermal energy (relative): {steady_energy:.2f}")

print(f"\n✓ Animation ready! Close the plot window to end the program.")
print(f"✓ The animation shows heat diffusion from the boundaries toward the center.")
print(f"✓ Notice how the temperature gradient decreases over time as steady state is approached.")

# Show the animation
plt.show()

print("\nSimulation completed successfully!")
print("Key observations:")
print("- Heat flows from the hot end (100°C) to the cold end (0°C)")
print("- The center of the rod gradually heats up from room temperature")
print("- The temperature profile becomes linear at steady state")
print("- The process takes several hours due to steel's thermal properties")
```
