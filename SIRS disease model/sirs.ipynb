# Complete Guide to the SIRS Disease Model

## Introduction

The SIRS model is a mathematical framework used to understand how infectious diseases spread through populations. It's an extension of the simpler SIR model that accounts for the loss of immunity over time. These models help epidemiologists predict disease outbreaks, evaluate intervention strategies, and understand disease dynamics.

## What is a Compartmental Model?

Compartmental models divide a population into distinct groups (compartments) based on disease status. Individuals move between these compartments as their disease status changes over time. Think of it like water flowing between connected tanks - people flow from one compartment to another at rates determined by the disease characteristics.

## The SIRS Compartments

The SIRS model divides the population into four compartments:

- **S (Susceptible)**: Individuals who can catch the disease
- **I (Infected)**: Individuals who have the disease and can transmit it to others
- **R (Recovered)**: Individuals who have recovered and are temporarily immune
- **S (Susceptible again)**: Recovered individuals return to susceptible status when immunity wanes

## Key Parameters and Their Meanings

Before deriving the equations, we need to understand the parameters that govern disease transmission:

- $N$: Total population size (assumed constant)
- $\beta$: Transmission rate (how easily the disease spreads)
- $\gamma$: Recovery rate (how quickly infected individuals recover)
- $\omega$: Rate of immunity loss (how quickly recovered individuals become susceptible again)

### Understanding the Transmission Rate $\beta$

The parameter $\beta$ represents the average number of contacts per unit time between an infected individual and susceptible individuals, multiplied by the probability that such contact results in transmission. 

If an infected person makes $c$ contacts per day and each contact has probability $p$ of transmission, then $\beta = cp$.

### Understanding the Recovery Rate $\gamma$

The recovery rate $\gamma$ is the reciprocal of the average infectious period. If individuals are infectious for an average of $D$ days, then $\gamma = 1/D$.

### Understanding the Immunity Loss Rate $\omega$

The immunity loss rate $\omega$ is the reciprocal of the average duration of immunity. If immunity lasts an average of $L$ time units, then $\omega = 1/L$.

## Key Assumptions

The SIRS model makes several important assumptions:

1. **Homogeneous mixing**: Every individual has an equal chance of encountering any other individual
2. **Constant population**: Birth and death rates are equal (or both are zero)
3. **No disease-induced mortality**: The disease doesn't kill anyone
4. **Uniform recovery**: All infected individuals recover at the same average rate
5. **Temporary immunity**: Recovered individuals eventually lose immunity and become susceptible again

## Derivation from First Principles

### The Fundamental Principle

The derivation is based on a fundamental principle from physics and chemistry: **conservation**. For each compartment, the rate of change equals the rate at which individuals enter minus the rate at which they leave.

$$\frac{d(\text{Compartment})}{dt} = \text{Rate In} - \text{Rate Out}$$

This principle comes from the continuity equation in physics, which states that mass (or in our case, people) cannot appear or disappear spontaneously.

### Analyzing Each Compartment

Let's denote:
- $S(t)$ = number of susceptible individuals at time $t$
- $I(t)$ = number of infected individuals at time $t$  
- $R(t)$ = number of recovered individuals at time $t$

#### For the Susceptible Compartment

**What flows in?** Recovered individuals who lose immunity
- Rate in = $\omega R(t)$

**What flows out?** Susceptible individuals who become infected
- To calculate this, we need to determine how many new infections occur per unit time
- An infected individual makes contacts at rate $\beta$
- The fraction of contacts that are with susceptible people is $S(t)/N$
- So each infected person causes $\beta S(t)/N$ new infections per unit time
- With $I(t)$ infected individuals, the total infection rate is $\beta S(t)I(t)/N$
- Rate out = $\beta S(t)I(t)/N$

Therefore:
$$\frac{dS}{dt} = \omega R - \beta \frac{SI}{N}$$

#### For the Infected Compartment

**What flows in?** Susceptible individuals who become infected
- Rate in = $\beta SI/N$

**What flows out?** Infected individuals who recover
- Rate out = $\gamma I(t)$

Therefore:
$$\frac{dI}{dt} = \beta \frac{SI}{N} - \gamma I$$

#### For the Recovered Compartment

**What flows in?** Infected individuals who recover
- Rate in = $\gamma I(t)$

**What flows out?** Recovered individuals who lose immunity
- Rate out = $\omega R(t)$

Therefore:
$$\frac{dR}{dt} = \gamma I - \omega R$$

## The Complete SIRS System

The complete system of differential equations is:

$$\begin{align}
\frac{dS}{dt} &= \omega R - \beta \frac{SI}{N} \\
\frac{dI}{dt} &= \beta \frac{SI}{N} - \gamma I \\
\frac{dR}{dt} &= \gamma I - \omega R
\end{align}$$

### Conservation Check

Since the total population is constant, we should have:
$$\frac{d}{dt}(S + I + R) = 0$$

Let's verify:
$$\frac{dS}{dt} + \frac{dI}{dt} + \frac{dR}{dt} = \left(\omega R - \beta \frac{SI}{N}\right) + \left(\beta \frac{SI}{N} - \gamma I\right) + \left(\gamma I - \omega R\right) = 0$$

The terms cancel out perfectly, confirming our equations are consistent with population conservation.

## The Basic Reproduction Number

An important quantity in epidemiology is the basic reproduction number $R_0$, which represents the average number of secondary infections caused by one infected individual in a completely susceptible population.

For the SIRS model:
$$R_0 = \frac{\beta}{\gamma}$$

This comes from the fact that an infected individual:
- Remains infectious for an average time of $1/\gamma$
- Makes $\beta$ effective contacts per unit time
- Therefore causes $\beta \times (1/\gamma) = \beta/\gamma$ infections on average

## Relationship to the SIR Model

The classic SIR model is a special case of SIRS where $\omega = 0$ (immunity is permanent). In this case, the system becomes:

$$\begin{align}
\frac{dS}{dt} &= - \beta \frac{SI}{N} \\
\frac{dI}{dt} &= \beta \frac{SI}{N} - \gamma I \\
\frac{dR}{dt} &= \gamma I
\end{align}$$

The SIRS model is particularly relevant for diseases where immunity wanes over time, such as influenza, common cold, or some bacterial infections.

## Equilibrium Analysis

Unlike the SIR model, the SIRS model can reach a steady state where the disease persists in the population. At equilibrium, all derivatives equal zero:

$$\frac{dS}{dt} = \frac{dI}{dt} = \frac{dR}{dt} = 0$$

This leads to an endemic equilibrium where the disease continues to circulate in the population at constant levels, sustained by the continuous loss of immunity in recovered individuals.

## Applications and Extensions

The SIRS model provides insights into:
- Endemic disease patterns
- The effect of vaccination campaigns (reducing $S$)
- The impact of immunity duration on disease persistence
- Seasonal disease patterns (when parameters vary with time)

More complex models can include births and deaths, multiple disease strains, spatial structure, or age-dependent parameters, but they all build on the fundamental principles demonstrated in the SIRS model.

## Visualizing the Model: From Theory to Simulation

Now that we understand the mathematical foundation of the SIRS model, the next step is to see it in action. Mathematical models truly come alive when we can visualize their behavior over time. 

The system of differential equations we derived:
$\begin{align}
\frac{dS}{dt} &= \omega R - \beta \frac{SI}{N} \\
\frac{dI}{dt} &= \beta \frac{SI}{N} - \gamma I \\
\frac{dR}{dt} &= \gamma I - \omega R
\end{align}$

describes how the populations change continuously over time, but to see these changes, we need to solve the equations numerically and plot the results.

Python provides excellent tools for this task. We can use:
- **scipy.integrate.odeint** to solve the differential equations numerically
- **matplotlib** to create dynamic animations showing how the disease spreads
- **numpy** for efficient numerical computations

The animation will show us several key behaviors:
- How quickly the disease spreads initially (the epidemic phase)
- Whether the disease dies out or reaches an endemic steady state
- The cyclical patterns that can emerge when immunity wanes
- How different parameter values affect the overall dynamics

This visualization transforms abstract mathematical concepts into intuitive understanding of disease dynamics.

## Summary

The SIRS model demonstrates how mathematical thinking can capture the essential dynamics of disease transmission. By applying the simple principle that the rate of change equals inflow minus outflow to each population compartment, we derived a system of differential equations that can predict complex epidemic behaviors. The key insight is that when immunity is temporary, diseases can persist indefinitely in populations, cycling between periods of high and low prevalence.

The mathematical framework we've developed provides the foundation for understanding more complex epidemiological models and informs public health decision-making. In the accompanying Python implementation, we'll see these mathematical concepts come to life through dynamic visualization.

```python

"""
SIRS Disease Model Animation

This script implements and animates the SIRS (Susceptible-Infected-Recovered-Susceptible) 
disease model, showing how infectious diseases spread through populations when immunity wanes.

The model is governed by the system of differential equations:
    dS/dt = ω*R - β*S*I/N
    dI/dt = β*S*I/N - γ*I  
    dR/dt = γ*I - ω*R
"""

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from scipy.integrate import odeint

class SIRSModel:
    """
    SIRS Disease Model Implementation
    
    Parameters:
    -----------
    beta : float
        Transmission rate (contacts per day × probability of transmission)
    gamma : float  
        Recovery rate (1/infectious_period)
    omega : float
        Immunity loss rate (1/immunity_duration)
    N : int
        Total population size
    """
    
    def __init__(self, beta=0.3, gamma=0.1, omega=0.01, N=1000):
        self.beta = beta    # Transmission rate
        self.gamma = gamma  # Recovery rate  
        self.omega = omega  # Immunity loss rate
        self.N = N         # Total population
        
        # Calculate basic reproduction number
        self.R0 = beta / gamma
        
        # Calculate average periods
        self.infectious_period = 1 / gamma
        self.immunity_duration = 1 / omega
        
    def derivatives(self, y, t):
        """
        Calculate derivatives for the SIRS model
        
        Parameters:
        -----------
        y : array_like
            Current values [S, I, R]
        t : float
            Current time
            
        Returns:
        --------
        dydt : array_like
            Derivatives [dS/dt, dI/dt, dR/dt]
        """
        S, I, R = y
        
        # Calculate the force of infection
        force_of_infection = self.beta * S * I / self.N
        
        # System of differential equations
        dS_dt = self.omega * R - force_of_infection
        dI_dt = force_of_infection - self.gamma * I
        dR_dt = self.gamma * I - self.omega * R
        
        return [dS_dt, dI_dt, dR_dt]
    
    def solve(self, initial_conditions, time_points):
        """
        Solve the SIRS model equations
        
        Parameters:
        -----------
        initial_conditions : array_like
            Initial values [S0, I0, R0]
        time_points : array_like
            Time points for solution
            
        Returns:
        --------
        solution : ndarray
            Solution array with shape (len(time_points), 3)
        """
        return odeint(self.derivatives, initial_conditions, time_points)

def create_animation():
    """
    Create an animated visualization of the SIRS model
    """
    # Model parameters
    model = SIRSModel(beta=0.4, gamma=0.1, omega=0.02, N=1000)
    
    # Initial conditions: mostly susceptible, few infected, none recovered
    S0, I0, R0 = 995, 5, 0
    initial_conditions = [S0, I0, R0]
    
    # Time points
    t_max = 200  # days
    t = np.linspace(0, t_max, 1000)
    
    # Solve the model
    solution = model.solve(initial_conditions, t)
    S, I, R = solution.T
    
    # Set up the figure and axes
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))
    fig.suptitle('SIRS Disease Model Animation', fontsize=16, fontweight='bold')
    
    # Main plot
    ax1.set_xlim(0, t_max)
    ax1.set_ylim(0, model.N)
    ax1.set_xlabel('Time (days)')
    ax1.set_ylabel('Number of Individuals')
    ax1.grid(True, alpha=0.3)
    ax1.legend(loc='upper right')
    
    # Phase plot (I vs S)
    ax2.set_xlim(0, model.N)
    ax2.set_ylim(0, max(I) * 1.1)
    ax2.set_xlabel('Susceptible (S)')
    ax2.set_ylabel('Infected (I)')
    ax2.set_title('Phase Portrait: Infected vs Susceptible')
    ax2.grid(True, alpha=0.3)
    
    # Initialize empty line objects
    line_s, = ax1.plot([], [], 'b-', linewidth=2, label='Susceptible')
    line_i, = ax1.plot([], [], 'r-', linewidth=2, label='Infected') 
    line_r, = ax1.plot([], [], 'g-', linewidth=2, label='Recovered')
    
    phase_line, = ax2.plot([], [], 'purple', linewidth=2, alpha=0.7)
    phase_point, = ax2.plot([], [], 'ro', markersize=8)
    
    # Add parameter information
    param_text = (f'Parameters:\n'
                 f'β = {model.beta:.3f} (transmission rate)\n'
                 f'γ = {model.gamma:.3f} (recovery rate)\n'
                 f'ω = {model.omega:.3f} (immunity loss rate)\n'
                 f'R₀ = {model.R0:.2f} (basic reproduction number)\n'
                 f'Infectious period = {model.infectious_period:.1f} days\n'
                 f'Immunity duration = {model.immunity_duration:.1f} days\n'
                 f'Population = {model.N}')
    
    ax1.text(0.02, 0.98, param_text, transform=ax1.transAxes, 
             verticalalignment='top', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.8),
             fontfamily='monospace', fontsize=9)
    
    def animate(frame):
        """Animation function"""
        # Current time index
        current_idx = min(frame * 2, len(t) - 1)  # Slow down animation
        
        # Update main plot
        line_s.set_data(t[:current_idx], S[:current_idx])
        line_i.set_data(t[:current_idx], I[:current_idx])
        line_r.set_data(t[:current_idx], R[:current_idx])
        
        # Update phase plot
        phase_line.set_data(S[:current_idx], I[:current_idx])
        if current_idx > 0:
            phase_point.set_data([S[current_idx-1]], [I[current_idx-1]])
        
        # Update title with current time
        current_time = t[current_idx]
        ax1.set_title(f'SIRS Model Dynamics (t = {current_time:.1f} days)', fontsize=12)
        
        return line_s, line_i, line_r, phase_line, phase_point
    
    # Create animation
    frames = len(t) // 2  # Reduce number of frames for smoother animation
    anim = animation.FuncAnimation(fig, animate, frames=frames, 
                                 interval=50, blit=True, repeat=True)
    
    plt.tight_layout()
    return fig, anim

def plot_comparison():
    """
    Create a comparison plot showing different scenarios
    """
    fig, axes = plt.subplots(2, 2, figsize=(15, 10))
    fig.suptitle('SIRS Model: Impact of Different Parameters', fontsize=16, fontweight='bold')
    
    # Time points
    t = np.linspace(0, 300, 1000)
    initial_conditions = [995, 5, 0]
    
    scenarios = [
        {'title': 'High Transmission (β=0.5)', 'params': {'beta': 0.5, 'gamma': 0.1, 'omega': 0.02}},
        {'title': 'Low Transmission (β=0.2)', 'params': {'beta': 0.2, 'gamma': 0.1, 'omega': 0.02}},
        {'title': 'Fast Immunity Loss (ω=0.05)', 'params': {'beta': 0.3, 'gamma': 0.1, 'omega': 0.05}},
        {'title': 'Slow Immunity Loss (ω=0.005)', 'params': {'beta': 0.3, 'gamma': 0.1, 'omega': 0.005}}
    ]
    
    for i, scenario in enumerate(scenarios):
        ax = axes[i//2, i%2]
        model = SIRSModel(**scenario['params'])
        solution = model.solve(initial_conditions, t)
        S, I, R = solution.T
        
        ax.plot(t, S, 'b-', label='Susceptible', linewidth=2)
        ax.plot(t, I, 'r-', label='Infected', linewidth=2)
        ax.plot(t, R, 'g-', label='Recovered', linewidth=2)
        
        ax.set_title(f"{scenario['title']} (R₀={model.R0:.2f})")
        ax.set_xlabel('Time (days)')
        ax.set_ylabel('Number of Individuals')
        ax.legend()
        ax.grid(True, alpha=0.3)
        ax.set_ylim(0, 1000)
    
    plt.tight_layout()
    return fig

def main():
    """
    Main function to run the SIRS model visualization
    """
    print("SIRS Disease Model Visualization")
    print("="*40)
    print("This script demonstrates the SIRS model with:")
    print("- Animated time series showing disease dynamics")
    print("- Phase portrait showing the relationship between S and I")
    print("- Parameter comparison showing different scenarios")
    print()
    
    # Create animated visualization
    print("Creating animated visualization...")
    fig_anim, anim = create_animation()
    
    # Create comparison plots
    print("Creating parameter comparison...")
    fig_comp = plot_comparison()
    
    # Show plots
    plt.show()
    
    print("\nKey Observations:")
    print("1. The SIRS model can show sustained oscillations due to immunity loss")
    print("2. Higher transmission rates lead to larger outbreaks")
    print("3. Faster immunity loss leads to more frequent epidemic cycles")
    print("4. The phase portrait shows the trajectory in the S-I plane")

if __name__ == "__main__":
    main()
```
