# A Beginner's Guide to Drone Vertical Flight Dynamics

## What is Drone Flight Dynamics?

Drone flight dynamics is the study of how forces and moments act on a drone to produce motion through three-dimensional space. Think of it as the physics that governs how a drone moves, rotates, and responds to control inputs.

A drone's complete motion can be broken down into six degrees of freedom:
- **Translation**: Movement along three axes (x, y, z)
- **Rotation**: Rotation around three axes (roll, pitch, yaw)

Understanding flight dynamics helps us:
- Design stable drones
- Create effective control systems
- Predict how a drone will behave
- Optimize performance and efficiency

## What is Vertical Flight Dynamics?

Vertical flight dynamics focuses specifically on the **up-and-down motion** of a drone along the vertical (z) axis. This is the simplest form of drone motion to understand and model.

When we study vertical flight dynamics, we're asking questions like:
- How much thrust do we need to hover?
- How fast will the drone climb with a given throttle input?
- What happens when we suddenly change the thrust?

## Mathematical Model Derivation

### Starting from First Principles

Let's build our mathematical model using Newton's fundamental laws of motion. We'll start simple and build complexity gradually.

### Step 1: Define Our System

Consider a drone of mass `m` moving only in the vertical direction (z-axis). We'll use a coordinate system where:
- **z** = altitude (positive upward)
- **ż** = vertical velocity (positive upward)  
- **z̈** = vertical acceleration (positive upward)

### Step 2: Identify the Forces

In vertical flight, two main forces act on our drone:

1. **Thrust Force (T)**: Generated by the rotors, acts upward
2. **Gravitational Force (W)**: Weight of the drone, acts downward

### Step 3: Apply Newton's Second Law

Newton's second law states: **ΣF = ma**

The net force in the vertical direction equals mass times acceleration:

```
Net Force = Thrust - Weight
ΣF_z = T - mg
```

Applying Newton's second law:
```
T - mg = mz̈
```

### Step 4: Solve for Acceleration

Rearranging to find vertical acceleration:

```
z̈ = (T - mg)/m
```

This can also be written as:
```
z̈ = T/m - g
```

### Step 5: The Complete Vertical Dynamics Equation

Our fundamental equation for vertical flight dynamics is:

```
mz̈ = T - mg
```

Or in acceleration form:
```
z̈ = (T - mg)/m
```

## Understanding the Equation

Let's analyze what this equation tells us:

### Case 1: Hovering (z̈ = 0)
For the drone to hover (no acceleration), we need:
```
0 = (T - mg)/m
T = mg
```
**The thrust must exactly equal the weight.**

### Case 2: Climbing (z̈ > 0)
For the drone to climb upward:
```
T > mg
```
**Thrust must be greater than weight.**

### Case 3: Descending (z̈ < 0)
For the drone to descend:
```
T < mg
```
**Thrust must be less than weight.**

## Relating Thrust to Control Input

In practice, thrust is controlled by the throttle input. For a simple model:

```
T = k_t × throttle_input
```

Where `k_t` is the thrust coefficient (thrust per unit throttle).

Substituting into our main equation:
```
z̈ = (k_t × throttle_input - mg)/m
```

## Example: Numerical Analysis

Let's say we have a drone with:
- Mass: m = 2 kg
- Thrust coefficient: k_t = 30 N per unit throttle
- Gravity: g = 9.81 m/s²

**For hovering:**
```
T_hover = mg = 2 × 9.81 = 19.62 N
throttle_hover = T_hover/k_t = 19.62/30 = 0.654
```

**For climbing at 2 m/s²:**
```
T_climb = m(g + z̈) = 2(9.81 + 2) = 23.62 N
throttle_climb = 23.62/30 = 0.787
```

## State-Space Representation

For control system design, we often express the dynamics in state-space form:

**States**: 
- x₁ = z (position)
- x₂ = ż (velocity)

**State equations**:
```
ẋ₁ = x₂
ẋ₂ = (T - mg)/m
```

**In matrix form**:
```
[ẋ₁]   [0  1] [x₁]   [0]
[ẋ₂] = [0  0] [x₂] + [1/m] T + [0]
                                [-g]
```

## Model Assumptions

Our simplified vertical flight dynamics model makes several key assumptions:

### Physical Assumptions
1. **Point Mass**: The drone is treated as a single point with mass concentrated at its center
2. **Rigid Body**: No structural flexibility or deformation
3. **Constant Mass**: Fuel consumption and battery weight changes are ignored
4. **Perfect Vertical Motion**: No coupling with horizontal motion or rotation

### Environmental Assumptions
5. **No Air Resistance**: Drag forces are neglected
6. **Constant Gravity**: Gravitational acceleration remains constant (valid for low altitudes)
7. **No Wind**: No external air currents or disturbances
8. **Standard Atmosphere**: Air density effects on rotor performance ignored

### Actuator Assumptions
9. **Instantaneous Thrust Response**: No rotor dynamics or time delays
10. **Linear Thrust Model**: Thrust is directly proportional to throttle input
11. **Perfect Thrust Control**: No actuator saturation or nonlinearities

### Sensor and Control Assumptions
12. **Perfect State Knowledge**: Position and velocity are known exactly
13. **No Measurement Noise**: Sensor readings are perfect
14. **No Control Delays**: Commands are executed instantly

## When This Model Applies

This simplified model works well for:
- Basic altitude control systems
- Initial drone design calculations
- Educational purposes and concept understanding
- Slow, gentle vertical maneuvers
- Small altitude changes where air density is constant

## Limitations and Next Steps

While this model captures the essential physics of vertical flight, real-world applications may require consideration of:
- Air resistance (especially at higher speeds)
- Rotor dynamics and time delays
- Ground effect when near the surface
- Atmospheric variations with altitude
- Coupling between vertical and horizontal motion
- Battery voltage effects on thrust capability

Understanding these fundamentals provides the foundation for more advanced models that account for these additional complexities.

## Bringing the Model to Life

Now that we have our mathematical foundation, the next logical step is to **implement and visualize** this model in action. A simulation will help us:

- See how the drone responds to different thrust inputs
- Understand the relationship between throttle commands and motion
- Experiment with different scenarios (hovering, climbing, descending)
- Validate our mathematical understanding through visual feedback

The animation will solve our governing equation **z̈ = (T - mg)/m** numerically and show the drone's position, velocity, and thrust in real-time. This bridges the gap between theory and practical understanding, making the abstract mathematics tangible and intuitive.

*Let's implement this model to see our vertical flight dynamics come alive!*

````python
"""
Drone Vertical Flight Dynamics Animation
========================================

This script animates the vertical flight dynamics model derived from first principles.
Governing equation: z̈ = (T - mg)/m
"""

import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.patches import Rectangle

class DroneVerticalDynamics:
    """
    Simulates vertical flight dynamics of a drone using the governing equation:
    z̈ = (T - mg)/m
    """
    
    def __init__(self, mass=2.0, thrust_coefficient=30.0, gravity=9.81):
        """
        Initialize drone parameters
        
        Args:
            mass (float): Drone mass in kg
            thrust_coefficient (float): Thrust per unit throttle (N)
            gravity (float): Gravitational acceleration (m/s²)
        """
        self.m = mass  # kg
        self.k_t = thrust_coefficient  # N per unit throttle
        self.g = gravity  # m/s²
        
        # Calculate hover throttle for reference
        self.hover_throttle = (self.m * self.g) / self.k_t
        
        # State variables [position, velocity]
        self.reset_state()
        
    def reset_state(self, initial_altitude=0.0, initial_velocity=0.0):
        """Reset drone to initial state"""
        self.z = initial_altitude  # altitude (m)
        self.z_dot = initial_velocity  # velocity (m/s)
        
    def get_thrust_from_throttle(self, throttle):
        """Convert throttle input (0-1) to thrust force (N)"""
        return self.k_t * throttle
    
    def dynamics(self, throttle):
        """
        Calculate acceleration using governing equation: z̈ = (T - mg)/m
        
        Args:
            throttle (float): Throttle input (0-1)
            
        Returns:
            float: Vertical acceleration (m/s²)
        """
        thrust = self.get_thrust_from_throttle(throttle)
        acceleration = (thrust - self.m * self.g) / self.m
        return acceleration
    
    def update(self, throttle, dt):
        """
        Update drone state using Euler integration
        
        Args:
            throttle (float): Current throttle input
            dt (float): Time step
        """
        # Calculate acceleration from current throttle
        z_ddot = self.dynamics(throttle)
        
        # Integrate using Euler's method
        self.z_dot += z_ddot * dt  # velocity update
        self.z += self.z_dot * dt  # position update
        
        # Prevent drone from going below ground
        if self.z < 0:
            self.z = 0
            self.z_dot = max(0, self.z_dot)  # prevent bouncing into ground

def create_flight_scenario():
    """
    Create a realistic flight scenario demonstrating proper hovering:
    - Climb with T > mg (positive acceleration)
    - Decelerate with T < mg to reach zero velocity
    - True hover with T = mg at zero velocity (constant altitude)
    """
    # Time parameters
    total_time = 40.0  # seconds - longer to show proper phases
    dt = 0.05  # smaller time step for better accuracy
    time_steps = int(total_time / dt)
    
    # Initialize arrays
    time_array = np.linspace(0, total_time, time_steps)
    throttle_array = np.zeros(time_steps)
    
    # Create flight phases
    drone = DroneVerticalDynamics()
    
    # Debug: Print hover throttle verification
    print(f"=== HOVER THRUST VERIFICATION ===")
    print(f"Drone mass (m): {drone.m} kg")
    print(f"Gravity (g): {drone.g} m/s²")
    print(f"Weight (mg): {drone.m * drone.g:.2f} N")
    print(f"Thrust coefficient (k_t): {drone.k_t} N per unit throttle")
    print(f"Calculated hover throttle: {drone.hover_throttle:.4f}")
    print(f"Hover thrust at this throttle: {drone.get_thrust_from_throttle(drone.hover_throttle):.2f} N")
    print(f"Expected acceleration at hover: {drone.dynamics(drone.hover_throttle):.6f} m/s² (should be ≈ 0)")
    print()
    
    for i, t in enumerate(time_array):
        if t < 3.0:
            # Phase 1: Start at rest - true hover at ground level
            throttle_array[i] = drone.hover_throttle
        elif t < 8.0:
            # Phase 2: Climb - accelerate upward
            throttle_array[i] = drone.hover_throttle * 1.3  # 30% more thrust
        elif t < 12.0:
            # Phase 3: Decelerate to stop climbing (brake phase)
            throttle_array[i] = drone.hover_throttle * 0.7  # 30% less thrust
        elif t < 18.0:
            # Phase 4: True hover - maintain constant altitude with zero velocity
            throttle_array[i] = drone.hover_throttle
        elif t < 23.0:
            # Phase 5: Descend - accelerate downward
            throttle_array[i] = drone.hover_throttle * 0.7  # 30% less thrust
        elif t < 27.0:
            # Phase 6: Decelerate to stop descending (brake phase)
            throttle_array[i] = drone.hover_throttle * 1.3  # 30% more thrust
        else:
            # Phase 7: Final true hover - maintain constant altitude
            throttle_array[i] = drone.hover_throttle
    
    return time_array, throttle_array

def animate_drone_flight():
    """Create and run the drone flight animation"""
    
    # Create flight scenario
    time_array, throttle_array = create_flight_scenario()
    dt = time_array[1] - time_array[0]
    
    # Initialize drone
    drone = DroneVerticalDynamics()
    
    # Simulate the entire flight to get data
    altitude_data = []
    velocity_data = []
    thrust_data = []
    acceleration_data = []
    
    # Debug: Print some key values during simulation
    print(f"Time step: {dt:.4f} seconds")
    print("\n=== SIMULATION PROGRESS ===")
    print("Key Principle: When T = mg, altitude should remain constant!")
    print("Time    Phase           Throttle  Thrust   Accel    Velocity  Altitude")
    print("(s)                     (-)       (N)      (m/s²)   (m/s)     (m)")
    print("-" * 75)
    
    drone.reset_state()
    for i, throttle in enumerate(throttle_array):
        altitude_data.append(drone.z)
        velocity_data.append(drone.z_dot)
        thrust = drone.get_thrust_from_throttle(throttle)
        thrust_data.append(thrust)
        accel = drone.dynamics(throttle)
        acceleration_data.append(accel)
        
        # Determine flight phase
        t = time_array[i]
        if t < 3.0:
            phase = "HOVER"
        elif t < 8.0:
            phase = "CLIMB"
        elif t < 12.0:
            phase = "BRAKE_UP"  # Decelerate to stop climbing
        elif t < 18.0:
            phase = "HOVER"
        elif t < 23.0:
            phase = "DESCEND"
        elif t < 27.0:
            phase = "BRAKE_DOWN"  # Decelerate to stop descending
        else:
            phase = "HOVER"
        
        # Print detailed info every 2 seconds
        if i % int(2.0 / dt) == 0 or t in [8.0, 12.0, 18.0, 23.0, 27.0]:  # Every 2 seconds + key transitions
            print(f"{t:4.1f}    {phase:<12}    {throttle:.4f}    {thrust:6.2f}   {accel:7.3f}    {drone.z_dot:7.3f}   {drone.z:7.3f}")
        
        drone.update(throttle, dt)
    
    # Convert to numpy arrays
    altitude_data = np.array(altitude_data)
    velocity_data = np.array(velocity_data)
    thrust_data = np.array(thrust_data)
    acceleration_data = np.array(acceleration_data)
    
    print(f"\n=== SIMULATION COMPLETE ===")
    print(f"Final altitude: {altitude_data[-1]:.3f}m")
    print(f"Max altitude: {np.max(altitude_data):.3f}m")
    print(f"Min altitude: {np.min(altitude_data):.3f}m")
    
    # Verify hover behavior
    print(f"\n=== HOVER BEHAVIOR VERIFICATION ===")
    print("Checking phases where TRUE hovering should occur...")
    hover_indices = []
    
    # Check true hover phases (where we expect both zero acceleration AND near-zero velocity)
    for i, t in enumerate(time_array):
        if (t < 3.0) or (12.0 <= t < 18.0) or (t >= 27.0):
            hover_indices.append(i)
    
    if hover_indices:
        hover_accelerations = acceleration_data[hover_indices]
        hover_velocities = velocity_data[hover_indices]
        
        max_hover_accel = np.max(np.abs(hover_accelerations))
        max_hover_velocity = np.max(np.abs(hover_velocities))
        
        print(f"Maximum acceleration during hover phases: {max_hover_accel:.6f} m/s²")
        print(f"Maximum velocity during hover phases: {max_hover_velocity:.6f} m/s")
        print(f"(Both should be very close to 0)")
        
        if max_hover_accel < 1e-6 and max_hover_velocity < 0.1:
            print("✅ PASS: True hover behavior is correct!")
        elif max_hover_accel < 1e-6:
            print("⚠️  PARTIAL: Acceleration is zero but velocity is not - check deceleration phases")
        else:
            print("❌ FAIL: Hover accelerations are larger than expected")
            
        # Show altitude stability during hover phases
        if len(hover_indices) > 1:
            hover_altitudes = altitude_data[hover_indices]
            altitude_variation = np.max(hover_altitudes) - np.min(hover_altitudes)
            print(f"Altitude variation during hover phases: {altitude_variation:.6f} m")
            if altitude_variation < 0.001:
                print("✅ PASS: Altitude remains constant during hover!")
    
    # Create the figure and subplots
    fig = plt.figure(figsize=(15, 10))
    fig.suptitle('Drone Vertical Flight Dynamics: z̈ = (T - mg)/m', fontsize=16, fontweight='bold')
    
    # Subplot 1: Drone visualization
    ax1 = plt.subplot(2, 3, (1, 4))
    ax1.set_xlim(-2, 2)
    ax1.set_ylim(-1, max(altitude_data) + 2)
    ax1.set_xlabel('Horizontal Position (m)')
    ax1.set_ylabel('Altitude (m)')
    ax1.set_title('Drone Position')
    ax1.grid(True, alpha=0.3)
    
    # Draw ground
    ground = Rectangle((-2, -0.5), 4, 0.5, facecolor='brown', alpha=0.7)
    ax1.add_patch(ground)
    
    # Drone representation
    drone_body = Rectangle((-0.3, 0), 0.6, 0.2, facecolor='blue', alpha=0.8)
    ax1.add_patch(drone_body)
    
    # Propellers
    prop_left = Rectangle((-0.5, 0.2), 0.15, 0.05, facecolor='gray')
    prop_right = Rectangle((0.35, 0.2), 0.15, 0.05, facecolor='gray')
    ax1.add_patch(prop_left)
    ax1.add_patch(prop_right)
    
    # Altitude line and text
    altitude_line, = ax1.plot([1.5, 1.5], [0, 0], 'r--', linewidth=2)
    altitude_text = ax1.text(1.7, 0, '0.0 m', fontsize=12, color='red')
    
    # Subplot 2: Altitude vs Time
    ax2 = plt.subplot(2, 3, 2)
    ax2.set_xlabel('Time (s)')
    ax2.set_ylabel('Altitude (m)')
    ax2.set_title('Altitude vs Time')
    ax2.grid(True, alpha=0.3)
    altitude_line_plot, = ax2.plot([], [], 'b-', linewidth=2, label='Altitude')
    ax2.legend()
    
    # Subplot 3: Velocity vs Time
    ax3 = plt.subplot(2, 3, 3)
    ax3.set_xlabel('Time (s)')
    ax3.set_ylabel('Velocity (m/s)')
    ax3.set_title('Velocity vs Time')
    ax3.grid(True, alpha=0.3)
    velocity_line_plot, = ax3.plot([], [], 'g-', linewidth=2, label='Velocity')
    ax3.legend()
    
    # Subplot 4: Thrust vs Time
    ax4 = plt.subplot(2, 3, 5)
    ax4.set_xlabel('Time (s)')
    ax4.set_ylabel('Thrust (N)')
    ax4.set_title('Thrust vs Time')
    ax4.grid(True, alpha=0.3)
    thrust_line_plot, = ax4.plot([], [], 'r-', linewidth=2, label='Thrust')
    hover_thrust_line = ax4.axhline(y=drone.m * drone.g, color='orange', 
                                   linestyle='--', label=f'Hover Thrust ({drone.m * drone.g:.1f} N)')
    ax4.legend()
    
    # Subplot 5: Acceleration vs Time
    ax5 = plt.subplot(2, 3, 6)
    ax5.set_xlabel('Time (s)')
    ax5.set_ylabel('Acceleration (m/s²)')
    ax5.set_title('Acceleration vs Time')
    ax5.grid(True, alpha=0.3)
    accel_line_plot, = ax5.plot([], [], 'm-', linewidth=2, label='Acceleration')
    zero_accel_line = ax5.axhline(y=0, color='black', linestyle='--', alpha=0.5, label='Zero Acceleration')
    ax5.legend()
    
    # Set plot limits
    ax2.set_xlim(0, time_array[-1])
    ax2.set_ylim(min(altitude_data) - 1, max(altitude_data) + 1)
    
    ax3.set_xlim(0, time_array[-1])
    ax3.set_ylim(min(velocity_data) - 1, max(velocity_data) + 1)
    
    ax4.set_xlim(0, time_array[-1])
    ax4.set_ylim(min(thrust_data) - 2, max(thrust_data) + 2)
    
    ax5.set_xlim(0, time_array[-1])
    ax5.set_ylim(min(acceleration_data) - 1, max(acceleration_data) + 1)
    
    # Animation function
    def animate(frame):
        # Update drone position
        current_altitude = altitude_data[frame]
        drone_body.set_y(current_altitude)
        prop_left.set_y(current_altitude + 0.2)
        prop_right.set_y(current_altitude + 0.2)
        
        # Update altitude indicator
        altitude_line.set_data([1.5, 1.5], [0, current_altitude])
        altitude_text.set_position((1.7, current_altitude / 2))
        altitude_text.set_text(f'{current_altitude:.1f} m')
        
        # Update time series plots
        current_time = time_array[:frame+1]
        
        altitude_line_plot.set_data(current_time, altitude_data[:frame+1])
        velocity_line_plot.set_data(current_time, velocity_data[:frame+1])
        thrust_line_plot.set_data(current_time, thrust_data[:frame+1])
        accel_line_plot.set_data(current_time, acceleration_data[:frame+1])
        
        return (drone_body, prop_left, prop_right, altitude_line, altitude_text,
                altitude_line_plot, velocity_line_plot, thrust_line_plot, accel_line_plot)
    
    # Create and run animation
    anim = animation.FuncAnimation(fig, animate, frames=len(time_array), 
                                 interval=50, blit=False, repeat=True)
    
    plt.tight_layout()
    plt.show()
    
    return anim

def print_model_info():
    """Print information about the model and simulation"""
    print("="*60)
    print("DRONE VERTICAL FLIGHT DYNAMICS SIMULATION")
    print("="*60)
    print("\nGoverning Equation: z̈ = (T - mg)/m")
    print("\nWhere:")
    print("  z̈ = vertical acceleration (m/s²)")
    print("  T = thrust force (N)")
    print("  m = drone mass (kg)")
    print("  g = gravitational acceleration (m/s²)")
    
    drone = DroneVerticalDynamics()
    print(f"\nDrone Parameters:")
    print(f"  Mass: {drone.m} kg")
    print(f"  Thrust Coefficient: {drone.k_t} N per unit throttle")
    print(f"  Gravity: {drone.g} m/s²")
    print(f"  Hover Throttle: {drone.hover_throttle:.3f}")
    print(f"  Hover Thrust: {drone.m * drone.g:.1f} N")
    
    print(f"\n=== FLIGHT SCENARIO ===")
    print("Demonstrating PROPER hovering behavior:")
    print("(True hover requires BOTH zero acceleration AND zero velocity)")
    print()
    print("  0-3s:   HOVER (T = mg, v = 0) → constant altitude")
    print("  3-8s:   CLIMB (T > mg) → positive acceleration, gaining upward velocity") 
    print("  8-12s:  BRAKE_UP (T < mg) → negative acceleration, reducing velocity to zero")
    print("  12-18s: HOVER (T = mg, v ≈ 0) → constant altitude (true hover!)")
    print("  18-23s: DESCEND (T < mg) → negative acceleration, gaining downward velocity")
    print("  23-27s: BRAKE_DOWN (T > mg) → positive acceleration, reducing velocity to zero")
    print("  27-40s: HOVER (T = mg, v ≈ 0) → constant altitude (true hover!)")
    print()
    print("Key Insight: Hover thrust (T = mg) only stops acceleration.")
    print("To achieve constant altitude, you must FIRST reduce velocity to zero!")
    print("Close the animation window to continue...")

if __name__ == "__main__":
    # Print model information
    print_model_info()
    
    # Run the animation
    anim = animate_drone_flight()
    
    print("\nAnimation complete!")
    print("Try modifying the flight scenario or drone parameters to see different behaviors!")
````
